```python
%%sql

drop table if exists sql_train.practice cascade;

create table sql_train.practice( 
    invoiceno integer     
    ,stockcode varchar(255)
    ,description varchar(255)
    ,quantity integer 
    ,invoicedate date    
    ,unitprice numeric 
    ,customerid integer 
);
copy
    sql_train.practice
from
    'C:\\Program Files\\PostgreSQL\\10\\data\\trn_data.csv'
with
csv
header
;
```

```python
%%sql
select
    *
FROM 
    sql_train.practice
limit 100;
```

人別どんな商品を買ったのか

```python
%%sql
select
    customerid
    , string_agg(stockcode,',')  as stockcodes
from
    sql_train.practice
group by 
    customerid
```

```python
%%sql
select
    regexp_split_to_table(description,' ') as product_id
from
    sql_train.practice
```

どんな言葉が多い？

```python
%%sql
with all_words as (
select
    regexp_split_to_table(description,' ') as words
from
    sql_train.practice
)
select
    words
    ,count(words) as count
from 
    all_words
group by 
    words
order by 
    count desc
```
