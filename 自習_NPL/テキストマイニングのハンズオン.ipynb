```python
file = open('text_fashion.txt',encoding='utf-8')
txt = file.read().replace('\n', '')
file.close()
```

```python
# テキストマイニング実行
t = Tokenizer()
tokens = t.tokenize(txt)
words = {}
```

```python

for token in tokens:
    data = str(token).split()[1].split(',')
    # 名詞のみ取得
    if data[0] == u'名詞':
        key = data[6]
        if key not in words:
            words[key] = 1
        else:
            words[key] += 1
```

```python
plt.rcParams['font.family'] = 'Yu Mincho'
plt.rcParams['font.family'] = 'IPAPGothic'
```
