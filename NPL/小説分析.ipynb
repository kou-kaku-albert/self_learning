```python
file = open(r'./king.txt')
txt = file.read().replace('\n', '')
file.close()
```

```python
txt
```

```python
from janome.tokenizer import Tokenizer

t = Tokenizer()
tokens = t.tokenize(txt)
words ={}
```

```python
for token in tokens:
    data=str(token).split()[1].split(',')
    print(token)
    print(data)
    if data[0] == u'名詞':
        key = data[6]
        if key not in words:
            words[key] = 1
        else:
            words[key] += 1
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
font_path=r"C:\Windows\Fonts\msgothic.ttc"
wc = WordCloud(font_path=font_path).generate_from_frequencies(words)
plt.figure(figsize=(20,20))
plt.imshow(wc)

```
